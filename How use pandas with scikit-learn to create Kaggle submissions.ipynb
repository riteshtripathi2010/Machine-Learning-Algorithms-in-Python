```python
import pandas as pd
train = pd.read_csv('data')
```

```python
train.head()
```

```python
#predict survival based on other characteristics
feat_cols = ['Pclass', 'Parch'] #parch: parents and children, i am here creating pythin list
X = train.loc[: , feat_cols]#this is the pandas dataframe that i have created
X.shape
```

```python
#lets create target factors
#the thing we are trying to predict, thats survival
y = train.Survived
y.shape
```

```python
#upto here, no need to convert, scikit will understand X and y as long as they are same dimensions

#lets introduce scikit learn
from sklearn.linear_model import LogisticRegression #this is a classification model

#lets instantiate the model
logreg = LogisticRegression()
logreg.fit(X,y)
```

```python
#lets read test data, where we predict data on
test = pd.read_csv('http://bit.ly/kaggletest')
test.head()#here missing column is the survived column
```

```python
#now lets create new X from testing data
X_new = test.loc[:, feat_cols]
X_new.shape
```

```python
#its 418 rows with 2 columns
#therefore we need to make 418 predictions

new_pred_class = logreg.predict(X_new)
```

```python
#now we need to put into csv file
#it wants passengerid and predictive classes
#test.PassengerId and new_pred_class
```

```python
pd.DataFrame({'PassengerId':test.PassengerId, 'Survived':new_pred_class})
```

```python
#remember dictionaries are unorder, it doesnt guarantee if passenger id will come first and we need it first coln
#to make it first coln:
pd.DataFrame({'PassengerId':test.PassengerId, 'Survived':new_pred_class}).set_index('PassengerId')
```

```python
#now to save it in csv file
pd.DataFrame({'PassengerId':test.PassengerId, 'Survived':new_pred_class}).set_index('PassengerId').to_csv('sub.csv')
```
