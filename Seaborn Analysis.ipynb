```python
#Seaborn
import seaborn as sns
df = sns.load_dataset("tips")
```

```python
df.head()#this the resuarant dataset
#we have to build a model, which will predict how much tip will be given
#tip is dependent feature and features are independent feature
```

```python
#Correlation and Heatmap
#corelation between each and every feature
df.corr()

#this will only work with integer features, values will be ranging between -1 to 1 and that is called pearson coefficeint corelation
#as total bill increasing, tip is also increasing

```

```python
sns.heatmap(df.corr())

#highly corelated is given by 1 
#we see tio and size are somewhere 0.5
#we use this for feature selection
#if both features are same, its better to use one feature
```

```python
#Join PLot: helps to do univarite ab=nalysis
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'hex')

#hex, shows in hexa form
#includes histogram
#many people have given tip of 2dollars and majority of bill is between 10 to 20
#there are some people who have given tip more than 9, bill of 50 and more, outlier it is

```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'reg')
#it will give probability definsity function
#it will draw regression line, best fit line
```

```python
#Now, if i have features more than two
#PAIR PLOT
sns.pairplot(df)
```

```python
#based on particular column / categorical value, eg: sex
sns.pairplot(df, hue = 'sex')
```

```python
#Number of smokers
df['smoker'].value_counts()
```

```python
#Dist Plot: helps you to create histograms
sns.distplot(df['tip'])#want to see how much people paid
#here 0.1, 0.2 is in percentage
```

```python
#i want to see another way as only histogram
sns.distplot(df['tip'], kde = False, bins = 10)
```

```python
#So far for categorical features: we have seen above

#Now, if i have categorical features
#what all plots we have
#box plot
#violin plot
#countplot
#bar plot
```

```python
#Count Plot
sns.countplot('sex', data = df)
```

```python
sns.countplot('day', data = df)#how people go out to eat on days
```

```python
sns.countplot(y = 'sex', data = df)
```

```python
#Bar Plot
sns.barplot(x = 'total_bill', y = 'sex', data = df)
```

```python
#in bar plots, u have to give x and y values
#count plot u have to give either x or y value
#incase of female, total bill was upto 17 
#i can use differetn features
```

```python
#Box Plot
#presents information from a five number summary
sns.boxplot('smoker', 'total_bill', data = df)
#the blue box lower indicates 0 %, 25%tile, 50 percentile, 75%tile and 100
```

```python
#smoker gave bill more than 50 as outlier and so on
```

```python
#using different colors
sns.boxplot(x = 'day', y = 'total_bill', data = df, palette = 'rainbow')
```

```python
#categorize my data based on some other categories
sns.boxplot(x = 'total_bill', y = 'day', hue = 'smoker', data = df)

#we can see different outliers on aprticualr days
```
